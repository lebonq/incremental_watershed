```python
import matplotlib.pyplot as plt
import numpy as np
import scienceplots
import pandas as pd

%matplotlib inline

latex_columnwidth_in_cm = 12.0
figure_aspect_ratio = 4/6

latex_columnwidth_in_pt = latex_columnwidth_in_cm/2.54

plt.style.use(['science','ieee']) # pip3 install SciencePlots --user

plt.rc('figure', dpi=150, figsize=[latex_columnwidth_in_pt, latex_columnwidth_in_pt*(4/6)])
plt.rc('font', size=8, serif='Times')
plt.rc('savefig', pad_inches=0.01)
```

## Coral

```python
data = pd.read_csv(r'/home/lebonq/CLionProjects/incremental_watershed/holiday_data/graph_coral.csv')
df = pd.DataFrame(data, columns=['IWS','OpenCV','HIGRA','NIWS','DIFT','Pixel Visited'])


fig, ax1 = plt.subplots()
x = np.linspace(0,25,26)
y = df[:]['IWS']
ax1.plot(x,y,label='$IWS$',color="#E6194B")
y = df[:]['NIWS']
ax1.plot(x,y,label='$NIWS$',color="#3CB44B")
y = df[:]['DIFT']
ax1.plot(x,y,label='$DIFT$',color="#911EB4")
y = df[:]['OpenCV']
ax1.plot(x,y,label='$OpenCV$',color="#0082C8")
y = df[:]['HIGRA']
ax1.plot(x,y,label='$Higra$',color="#F58231",linestyle="dashed")
ax1.set_yscale('log')
#plt.xlabel('\#interaction')
ax1.set_ylabel('Time in ms')

ax2 = ax1.twinx()
y = df[:]['Pixel Visited']
ax2.plot(x, y, label='Pixel Visited', color = '#000000')
ax2.tick_params(axis ='y', labelcolor = 'black')

plt.title("Coral")
ax2.grid(axis='x')
ax1.grid(axis='y')
plt.savefig('graph_coral.pdf', dpi=1000, transparent=True)
```

## Plant

```python
data = pd.read_csv(r'/home/lebonq/CLionProjects/incremental_watershed/holiday_data/graph_plant.csv')
df = pd.DataFrame(data, columns=['IWS','OpenCV','HIGRA','NIWS','DIFT','Pixel Visited'])

fig, ax1 = plt.subplots()
x = np.linspace(0,59,60)

y = df[:]['IWS']
line1, = ax1.plot(x,y,label='IWS',color="#E6194B")
y = df[:]['NIWS']
line2, = ax1.plot(x,y,label='NIWS',color="#3CB44B")
y = df[:]['DIFT']
line3, = ax1.plot(x,y,label='DIFT',color="#911EB4")
y = df[:]['OpenCV']
line4, = ax1.plot(x,y,label='OpenCV',color="#0082C8")
y = df[:]['HIGRA']
line5, = ax1.plot(x,y,label='Higra',color="#F58231",linestyle="dashed")
ax1.set_yscale('log')
#ax1.set_xlabel('\#interaction')
#ax1.set_ylabel('Time in ms')

ax2 = ax1.twinx()
y = df[:]['Pixel Visited']
line6, = ax2.plot(x, y, label='Visited pixels', color = 'black')
ax2.set_ylabel('Pixels', color = 'black')
ax2.tick_params(axis ='y', labelcolor = 'black')
ax2.grid(axis='x')
ax1.grid(axis='y')
fig.legend([line1, line2,line3,line4,line5,line6],["IWS","NIWS","DIFT","OpenCV","Higra","Visited pixels"],bbox_to_anchor=(0.91, 0.6), fontsize=6,handleheight=0.15)
plt.title("Plant")
plt.savefig('graph_plant.pdf', dpi=1000, transparent=True)
```
